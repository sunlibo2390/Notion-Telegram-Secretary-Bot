```python
import json
import pandas as pd

projects_path = r"D:\Projects\notion_secretary\databases\json\projects.json"
tasks_path = r"D:\Projects\notion_secretary\databases\json\tasks.json"
logs_path = r"D:\Projects\notion_secretary\databases\json\logs.json"
with open(projects_path, "r", encoding="utf-8") as f:
    projects_data = json.load(f)

with open(tasks_path, "r", encoding="utf-8") as f:
    tasks_data = json.load(f)

with open(logs_path, "r", encoding="utf-8") as f:
    logs_data = json.load(f)
```

```python
new_items = []
for item in projects_data['results']:
    new_item = {
        'name': item['properties']['Name']['title'][0]['plain_text'],
        'status': item['properties']['Status']['status']['name'],
        'id': item['id']
    }
    new_items.append(new_item)
projects_df = pd.DataFrame(new_items)
projects_df = projects_df[projects_df['status']!='Done']
projects_df
```

```python
from block_children_query import block_children_query
from block2md import notion_blocks_to_markdown
from tqdm import tqdm

new_items = []
for item in tqdm(tasks_data['results']):
    page_info = block_children_query(item['id'])
    md_text = notion_blocks_to_markdown(page_info['results'])
    new_item = {
        'name': item['properties']['Name']['title'][0]['plain_text'],
        'status': item['properties']['Status']['status']['name'],
        'id': item['id'],
        'content': md_text,
        'project_id': item['properties']['Projects']['relation'][0]['id'],
    }
    if new_item['status'] not in ['Done', 'Dormant']:
        new_items.append(new_item)
tasks_df = pd.DataFrame(new_items)
tasks_df
```

```python
from block_children_query import block_children_query
from block2md import notion_blocks_to_markdown
from tqdm import tqdm

new_items = []
for item in tqdm(logs_data['results']):
    page_info = block_children_query(item['id'])
    md_text = notion_blocks_to_markdown(page_info['results'])
    new_item = {
        'name': item['properties']['Name']['title'][0]['plain_text'],
        'status': item['properties']['Status']['status']['name'],
        'id': item['id'],
        'content': md_text,
        'task_id': item['properties']['Task']['relation'][0]['id'],
    }
    if new_item['status'] not in ['Done', 'Dormant']:
        new_items.append(new_item)
logs_df = pd.DataFrame(new_items)
logs_df
```
